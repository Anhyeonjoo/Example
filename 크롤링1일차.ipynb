# 네이버 뉴스기사 수집
- 뉴스제목, 내용 2가지 수집



```python
news_url = 'https://n.news.naver.com/mnews/article/215/0001045272?sid=103'

# 웹 페이지에서 강제 종료한 경우는 HTTP Request header를 설정해주면 해결 가능!
# 딕셔너리 구조로 생성해준다.
# key : value
header = {
    
    "user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36"
    
}

# header를 설정하는 이유는 서버측에서 컴퓨터프로그램이 요청한건지
# 브라우저가 요청한건지 판단을 할 수 있다! --> user-agent속성값으로 판단
# 요청할 때 이 값을 구성해서 요청하면 강제종료에 대한 문제를 해결할 수 있다.

res = req.get(news_url, headers=header) 
# headers 변수에 값 담아주기 /오류해결

html = bs(res.text,'lxml')
```

```python
# 뉴스제목, 내용을 수집한 후 출력해보기
#뉴스제목
title = html.select_one('h2.media_end_head_headline')
print(title.text)

```

```python
#뉴스내용 - id는 #을 써서 정의해준다.
# 공백지우는 함수 .strip() = 양쪽끝공백삭제해줌
content= html.select_one('#dic_area').text.strip()
print(content)
```

# 멜론 음원차트 수집



순위, 곡명, 아티스트, 앨범이미지주소 수집해서 가져오기

- csv 파일 저장


```python
import requests as req
import pandas as pd

from bs4 import BeautifulSoup as bs
```

```python
melon_url = 'https://www.melon.com/chart/'

#header 계속 이용 
header = {
    
    "user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36"
    
}

#요청
res= req.get(melon_url, headers = header)

html = bs(res.text,'lxml')
```

```python
# 순위 가져오기
ranks = html.select('td span.rank ')

# 곡명 가져오기
songNms = html.select('div.ellipsis.rank01 a')

# 아티스트 가져오기
artists = html.select('div.ellipsis.rank02 >span')

# 앨범이미지 주소 가져오기
albumimgs = html.select('a.image_typeAll > img')

# 크기가 같은지 확인해보기(여기선 100개 인지 확인)
len(ranks), len(songNms), len(artists), len(albumimgs)
```

```python
# 4개의 빈 리스트에 데이터를 저장해보기!(.append)
ranklist =[]
songNmlist=[]
artistslist =[]
albumimgslist =[]


# 여러개를 접근해야한다
#이때 반복하는건 길이로 해주는게 좋다
# 길이가 다 같으니까 아무거나 넣어줘도 된다.
for i in range(len(ranks)):
    ranklist.append(ranks[i].text)
    songNmlist.append(songNms[i].text)
    artistslist.append(artists[i].text)
    albumimgslist.append(albumimgs[i].get('src')) #이미지는 get('src') 사용해줘야함
    
```

```python
melon_dic = {
    
    '순위':ranklist,
    '곡명':songNmlist,
    '아티스트':artistslist,
    '앨범이미지':albumimgslist
    
}

melon_df = pd.DataFrame(melon_dic)
```

```python
melon_df
```

```python
# 저장할 각자의 확장자를 적어주면 된다.
# 지금 csv로 저장할거기 때문에 csv로 확장자 작성해준다.
# encoding안해주면 한글깨지기 때문에 해주자~!

melon_df.to_csv('melon_top100.csv', encoding='euc-kr', index=False)
#index를 false값으로 주면 0이 아닌 1부터 100까지 잘 나오는걸 확인가능하다!
```
